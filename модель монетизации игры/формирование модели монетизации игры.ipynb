### Какую стратегию предпочитают пользователи

Логика сбора данных не обепечивает возможности явно выделить пользователей, окончивших этап с помощью стратегии победы над другим игроком.
Добавим столбец с индикатором стратегии , где : 
 
 0 - в процессе прохождения
 
 1- окончил уровень с помощью победы
 
 2 - окончил уровеньс помощью постройки 

```python
# создадим список id клиентов, которые окончили уровень, с помощью стратегии постройки
list_project = actions[actions['event']=='project']['user_id'].unique()
# создадим список id клиентов, которые окончили уровень, с помощью стратегии победы
list_win = actions[(actions['event']=='finished_stage_1')
                   &(~actions['user_id'].isin(list_project))]['user_id'].unique()
           
```

```python
# проставим индикаторы стратегии
actions.loc[actions['user_id'].isin(list_win), 'strategy_type'] = 1
actions.loc[actions['user_id'].isin(list_project), 'strategy_type'] = 2
actions['strategy_type'] = actions['strategy_type'].fillna(0) 
actions['strategy_type'] = actions['strategy_type'].astype('int')
# создадим словарь индикаторов
dictionary = {0:'В процессе прохождения',1:'Победа первого противника ',2:'Реализация проекта'}
actions.head()
```

Посмотрим как распределяются игроки по типам стратегий.

```python
# группируем по типам стратегий и считаем количество и долю уникальных пользователей
strat_group=actions.groupby('strategy_type')['user_id'].nunique().to_frame().reset_index()
strat_group['ratio'] = (strat_group['user_id']/strat_group['user_id'].sum()*100).round(2)
# заменяем числовое значение стратегии на ее наименование
strat_group = strat_group.replace({'strategy_type':dictionary})

display(strat_group)

display("Общее количество игроков - {}".format(strat_group['user_id'].sum()))

# функция, проставляющая значения над барами
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i].round(1), y[i].round(1), ha = 'center',fontsize=15)
        
# визуализируем предпочтения игроков 
with plt.style.context('seaborn-pastel'):
    # столбчатая диаграмма
    plt.figure(figsize=(12, 8))
    plt.tick_params(labelsize=14)
    sns.barplot(data=strat_group,x='strategy_type',y='ratio')
    plt.title('Распределение игроков по стратегиям',fontsize=17)
    plt.xlabel('Стратегия',fontsize=12)
    addlabels(strat_group['strategy_type'],strat_group['ratio'])
    plt.ylabel('Доля(%)',fontsize=14)
    plt.xticks(rotation=0)
    plt.show
```

### Создание профилей пользователей

```python
# функция для создания пользовательских профилей

#def get_profiles(sessions, orders, events, ad_costs, event_names=[]):
def get_profiles(actions, user_source, costs):

    # находим параметры первых посещений
    profiles = (
        actions.merge(user_source,on='user_id',how='right')
        .sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg(
            {
                'event_datetime': 'first',
                #'event': 'first',
                'building_type': 'count',
                'strategy_type': 'first',
                'source': 'first',
            }
        )
        .rename(columns={'event_datetime': 'first_ts',
                        'building_type':'amount_building'})
        .reset_index()
    )

    # для когортного анализа определяем дату первого посещения
    # и первый день месяца, в который это посещение произошло
    profiles['day'] = profiles['first_ts'].dt.date#.astype('datetime64[M]') # было .dt.date
    profiles['day'] = pd.to_datetime(profiles['day'])
    profiles['month'] = profiles['first_ts'].astype('datetime64[M]')

    # считаем количество уникальных пользователей
    # с одинаковыми источником и датой привлечения
    new_users = (
        profiles.groupby(['day', 'source'])
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()
    )
    new_users['day'] = pd.to_datetime(new_users['day'])
    costs['day'] += dt.timedelta(days=1)
    # объединяем траты на рекламу и число привлечённых пользователей
    ad_costs = costs.merge(new_users, on=['day', 'source'], how='left')

    # делим рекламные расходы на число привлечённых пользователей
    ad_costs['acquisition_cost'] = ad_costs['cost'] / ad_costs['unique_users']

    # добавляем стоимость привлечения в профили
    profiles = profiles.merge(
        ad_costs[['day', 'source', 'acquisition_cost']],
        on=['day', 'source'],
        how='left',
    )
    # стоимость привлечения органических пользователей равна нулю
    profiles['acquisition_cost'] = profiles['acquisition_cost'].fillna(0)

    return profiles
```

```python
profiles = get_profiles(actions, user_source, costs)
profiles.head()
```

```python
# функция для расчёта LTV и ROI

def get_ltv(
    profiles,
    purchases_bas,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('day <= @last_suitable_acquisition_date')
    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        purchases_bas[['user_id', 'event_datetime', 'revenue']], on='user_id', how='left'
    )
    # рассчитываем лайфтайм пользователя для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_datetime'] - result_raw['first_ts']
    ).dt.days
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values='revenue', aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes

        # собираем датафрейм с данными пользователей и значениями CAC, 
        # добавляя параметры из dimensions
        cac = df[['user_id', 'acquisition_cost'] + dims].drop_duplicates()

        # считаем средний CAC по параметрам из dimensions
        cac = (
            cac.groupby(dims)
            .agg({'acquisition_cost': 'mean'})
            .rename(columns={'acquisition_cost': 'cac'})
        )

        # считаем ROI: делим LTV на CAC
        roi = result.div(cac['cac'], axis=0)

        # удаляем строки с бесконечным ROI
        roi = roi[~roi['cohort_size'].isin([np.inf])]

        # восстанавливаем размеры когорт в таблице ROI
        roi['cohort_size'] = cohort_sizes

        # добавляем CAC в таблицу ROI
        roi['cac'] = cac['cac']

        # в финальной таблице оставляем размеры когорт, CAC
        # и ROI в лайфтаймы, не превышающие горизонт анализа
        roi = roi[['cohort_size', 'cac'] + list(range(horizon_days))]

        # возвращаем таблицы LTV и ROI
        return result, roi

    # получаем таблицы LTV и ROI
    result_grouped, roi_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблиц динамики убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицы динамики LTV и ROI
    result_in_time, roi_in_time = group_by_dimensions(
        result_raw, dimensions + ['day'], horizon_days
    )

    return (
        result_raw,  # сырые данные
        result_grouped,  # таблица LTV
        result_in_time,  # таблица динамики LTV
        roi_grouped,  # таблица ROI
        roi_in_time,  # таблица динамики ROI
    ) 
```

```python
# функция для сглаживания фрейма

def filter_data(df, window):
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean() 
    return df 
```

```python
# функция для визуализации LTV и ROI

def plot_ltv_roi(ltv, ltv_history, roi, roi_history, horizon, window=1):

    # задаём сетку отрисовки графиков
    plt.figure(figsize=(20, 10))

    # из таблицы ltv исключаем размеры когорт
    ltv = ltv.drop(columns=['cohort_size'])
    # в таблице динамики ltv оставляем только нужный лайфтайм
    ltv_history = ltv_history.drop(columns=['cohort_size'])[[horizon - 1]]

    # стоимость привлечения запишем в отдельный фрейм
    cac_history = roi_history[['cac']]

    # из таблицы roi исключаем размеры когорт и cac
    roi = roi.drop(columns=['cohort_size', 'cac'])
    # в таблице динамики roi оставляем только нужный лайфтайм
    roi_history = roi_history.drop(columns=['cohort_size', 'cac'])[
        [horizon - 1]
    ]

    # первый график — кривые ltv
    ax1 = plt.subplot(2, 3, 1)
    ltv.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('LTV')

    # второй график — динамика ltv
    ax2 = plt.subplot(2, 3, 2, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in ltv_history.index.names if name not in ['day']]
    filtered_data = ltv_history.pivot_table(
        index='day', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика LTV пользователей на {}-й день'.format(horizon))

    # третий график — динамика cac
    ax3 = plt.subplot(2, 3, 3, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in cac_history.index.names if name not in ['day']]
    filtered_data = cac_history.pivot_table(
        index='day', columns=columns, values='cac', aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика стоимости привлечения пользователей')

    # четвёртый график — кривые roi
    ax4 = plt.subplot(2, 3, 4)
    roi.T.plot(grid=True, ax=ax4)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('ROI')

    # пятый график — динамика roi
    ax5 = plt.subplot(2, 3, 5, sharey=ax4)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in roi_history.index.names if name not in ['day']]
    filtered_data = roi_history.pivot_table(
        index='day', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax5)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.xlabel('Дата привлечения')
    plt.title('Динамика ROI пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show() 
```

```python
# момент анализа
observation_date = datetime(2020, 6, 5).date() 
# горизонт анализа 
horizon_days = 27 

# считаем LTV и ROI для показа рекламы при каждой постройке объекта
ltv_raw, ltv_grouped, ltv_history, roi_grouped, roi_history = get_ltv(
    profiles, purchases_bas, observation_date, horizon_days
)

# строим графики
plot_ltv_roi(ltv_grouped, ltv_history, roi_grouped, roi_history, horizon_days)
roi_grouped.T
```

## Проверка статистических гипотез

### Проверка  гипотезы о различии среднего количества построенных зданий 1 игроком разных стратегий

Перед проверкой гипотезы проанализируем данные на выбросы.

```python
# построим boxplot  количества зданий
sns.boxplot(profiles['amount_building'])
plt.show()
profiles.head()
```

Шумов в данных не обнаружено.

Проверим различаются ли средняя среднее количество построенных зданий 1 игроком двух разных стратегий.

Определим: 

  Нулевую гипотезу - Среднее количество построенных зданий не различается.

  Альтернативную  - Среднее количество построенных зданий  различается. 

Критический уровень статистической значимости зададим значением 0.05

```python
# количество построенных зданий игроками, завершающих уровень победой над противником
amount_building_1 = profiles.query('strategy_type==1')['amount_building']

# количество построенных зданий игроками, завершающих уровень проектом
amount_building_2 = profiles.query('strategy_type==2')['amount_building']  

#  зададим критический уровень статистической значимости
alpha = .05 

#  проверим гипотезу о равенстве средних двух независимых совокупностей
results = st.ttest_ind(amount_building_1,amount_building_2,equal_var = False)

print('p-значение:', results.pvalue)

# сравним  получившееся p-значение с заданным уровнем статистической значимости
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
print('Среднее количество объектов игрока, завершившего уровень с помощью победы' ,profiles.query('strategy_type==1')['amount_building'
                                                                                                ].mean())
print('Среднее количество объектов игрока, завершившего уровень проектом' ,profiles.query('strategy_type==2')['amount_building'
                                                                                                ].mean())
```

Полученное p-value в ходе проверки гипотезы говорит о том, что между средним двух выборок есть статистически значимое расхождение.

Среднее количество построенных зданий игроками разных стратегий отличается.Игроки, завершающие уровень с помощью реализации проекта строят больше зданий, в среднем 13 шт.,в то время как игроки другой стратегии строят 9 шт в среднем за игру.

Есть игроки, которые проходят уровень аномально долго, свыше 650 часов.Удалим таких игроков из выборки.

Определим: 

  Нулевую гипотезу - Среднее время прохождения уровня игроками разных стратегий не различается.

  Альтернативную  - Среднее время прохождения уровня игроками разных стратегий  различается. 

Критический уровень статистической значимости зададим значением 0.05

```python
# количество построенных зданий игроками, завершающих уровень победой над противником
time_level_1 = profiles.query('strategy_type==1')['diff']

# количество построенных зданий игроками, завершающих уровень проектом
time_level_2 = profiles.query('strategy_type==2')['diff']  

#  зададим критический уровень статистической значимости
alpha = .05 

#  проверим гипотезу о равенстве средних двух независимых совокупностей
results = st.ttest_ind(time_level_1,time_level_2,equal_var = False)

print('p-значение:', results.pvalue)

# сравним  получившееся p-значение с заданным уровнем статистической значимости
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

     4.Среднее количество построенных зданий одним игроком зависит от выбранной стратегии.
     
Игроки, выбирающие стратегию победы над противником, строят меньше зданий : в среднем -9,4шт. за всю игру.

Игроки , выбирающие другую стратегию строят большее количество зданий : в среднем -12,6 шт. за всю игру.
        
     5. Стратегия игрока влияет на время прохождение уровня 
     
Игроки, выбирающие стратегию победы над противником, тратят меньше времени на завершение уровня - в среднем, 11 дней . 

Игроки , выбирающие другую стратегию тратят 13,5 дней на завершение уровня.
          
При текущем поведении игроков возможно два варианта монетизации:
    
    - показ рекламы при каждой постройке здания,
    - показ рекламы, начиная со 2-го здания.
    
Оба варианта монетизации окупают  игроков.

Второй вариант также дает возможность игроку вовлечься в игру до начала показы рекламы. Это позволит сохранить лояльность пользователей и уменьшить их отток. 
 
В остальных случаях (если показывать рекламу реже) - стоимость привлечения игроков не покрывается доходами от рекламы.

Для того, чтобы показывать рекламу игрокам реже и сохранять их лояльность необходимо стимулировать игроков на более активную постройку зданий в рамках выбранной стратегии или выровнять предпочтения игроков по выбору стратегии. Если увеличить привлекательность стратегии завершения уровня с помощью реализации проекта, общее количество построек также увеличиться.

Снижение частоты показа рекламы может быть достигнута за счет оптимизации маркетинговых расходов.

Пользователи в независимости от канала привлечения демонстрируют однородность поведения.  
Однородность привлекаемых игроков позволяет сделать акцент на  наиболее дешевых каналах привлечения (youtube_channel_reklama и yandex_direct) без потери в доходах от монетизации. 
На данный момент маркетинговый бюджет распределен не оптимально , расходы на каналы примерно одинаковы , в то время как стоимость привлечения разная.

Проверка статистических гипотез показала дисбаланс в игре между 2-мя стратегиями, что требует внесения изменений в игровые механики . В противном случае стратегия прохождения уровня через реализацию проекта может выродиться.
