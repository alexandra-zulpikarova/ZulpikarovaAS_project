## Предобработка данных

### Обработка пропусков

Проверим нулевые значения количества комнат в стобце `rooms`: ошибочны ли данные или есть связь с другими характеристиками

```python
# построим сводную таблицу для квартир с нулевым значениям в rooms с группировкой по типу квартиры (студия или нет) и типу планировки
display(
    data.query('rooms == 0')
    .pivot_table(index='studio', columns='open_plan',values='rooms',aggfunc='count')
)
# посмотрим связь количества комнат в квартире и признака квартира-студия
display(data.groupby(['rooms','studio'])['rooms'].count())
# посмотрим связь количества комнат в квартире,площади квартиры и признака открытой планировки
data.groupby(['rooms','open_plan']).agg({'rooms':'count',
                                                  'total_area':'median'})
```

По результатам сводной таблицы видно, что для всех записей с нулевым значением комнат характерен или признак, что квартира - студия или признак открытой планировки. Видимо, возникновение нулевых значений в столбце `rooms` связано с тем, что нет единого очевидного подхода к заполнению количества комнат для квартир-студий и квартир с открытой планировкой.Кто-то из продавцов посчитал, что квартира-студия не имеет комнат и проставил "0", а кто посчитал, что в квартире-студии 1 комната, единая для всех жилых помещений.

Для корректного анализа необходимо данные о количестве комнат по квартирам-студиям и квартирам с открытой планировкой привести к единообразию. Будем считать ,что:
 - все квартиры-студии имеют 1 комнату;
 - квартиры с открытой планировкой имеют, как минимум, 1 комнату. Наличие большего количества комнат для таких квартир будет обусловлено ростом площади квартиры.В данном случае продавец хочет сказать  о том, сколько возможно разместить комнат в такой квартире или то , что изначально квартира была с открытой планировкой.

```python
#заменим все нулевые значения в столбце rooms на 1
data['rooms'] = data['rooms'].where(data['rooms'] >= 1, 1)
#проверим ,что не осталось нулевых значений в rooms 
print('Количество нулевых значений rooms после замены:',data.query('rooms == 0')['rooms'].count())
```

```python
# подсчёт пропусков
data.isna().sum()
```

```python
#посмотрим процентное содержание Nan в каждом столбце
for row in data:
    share = ((data[row].isna().sum())/len(data[row])*100).round(1)
    if share > 0:
        print(row,share)
```

Пропуски в `floors_total`,`locality_name` допустимо удалить , так как их доля менее 1%.

Возможно пропуски в `ceiling_height` связаны с тем, что продавец не знал точную высоту потолков в квартире при размещении объявления или считал данный параметр несущественным.Заменим пропуски на медианной значение высоты потолков.

 Для всех пропущенных значений `living_area` возможно заполнить пропуски расчетным значением. Определить отношение медианной  жилой площади к медианной общей площади в разрезе количества комнат и полученный коэффициент использовать для нахождения  пропущенных значений `living_area` , умножая его на соответсвущие значения `total_area`. 

Аналогично `living_area` заполним пропуски в  `kitchen_area`. 

Пропуски в `balcony` скорее всего связаны с отсутствием балконов в квартире, видимо люди пропускали заполнение столбца, если балконов в их квартире не было. Поэтому заменим пропуски в столбце на 0.

Доля пропусков в `days_exposition` - 13,4%. Удаление данных пропусков значительно сократить датасет и обьем информации по остальным показателям. Также некорректно заполнять данные пропуски каким-либо расчетным значением, так как отсутствие данных может объясняться  различными причинами. Это может быть как и продажа квартиры, по которой забыли снять объявление ,так и то, что квартира до сих пор не продана.Корректнее заполнить дефолтным значением,чтобы при дальнейшем исследовательском анализе была возможность сделать корректный срез данных.



Для всех пропусков , где не указано расстояние до центра , не указано и расстояние до ближайшего аэропорта  и `parks_around3000` и `ponds_around3000`,`parks_nearest`,`ponds_nearest`. 

Так как данные в`airports_nearest`,`cityCenters_nearest`,`parks_around3000`,`parks_nearest`,`ponds_around3000`,`ponds_nearest` получены автоматически на основе картографических данных, возможно пропуски в них связаны с тем,что продавцы данных квартир некорректно указали местоположение своей квартиры или по каким-либо причинам  не указали его, что не позволило программе рассчитать и подгрузить данные параметры.

Данные этих столбцов будет некорректно заполнять расчетными значениями или удалять ввиду большой доли пропусков от общего количества записей, поэтому будем заменять их служебным значением "999999", означающим отсутствие данных.



Удалим пропуски в `floors_total`,`locality_name`


```python
# применим функцию dropna к floors_total,locality_name
data = data.dropna(subset=['floors_total','locality_name'])
#проверим , что пропусков не осталось 
print('Количество пропусков в floors_total после обработки:',data['floors_total'].isna().sum())
print('Количество пропусков в locality_name после обработки:',data['locality_name'].isna().sum())
```

Обработаем пропуски в `is_apartment`


```python
#посмотрим количество уникальных значений is_apartment'
data['is_apartment'].value_counts()
```


Так как апартаменты достаточно редкий тип на рынке жилой недвижимости, предположим что пропущены значения относятся к False.Скорее всего люди пропускали заполнение этого столбца, если их квартиры не была апартаментами.

Заполним пропуски в столбце значением False


```python
#Заполним пропуски в столбце is_apartment значением False
data['is_apartment'] = data['is_apartment'].fillna(False)
#проверим , что пропусков не осталось 
print('Количество пропусков в is_apartment после обработки:',data['is_apartment'].isna().sum())
```

 Обработаем пропуски в столбце `balcony`.



```python
#Заполним пропуски в столбце balcony значением 0
data['balcony'] = data['balcony'].fillna(0)
#проверим , что пропусков не осталось 
print('Количество пропусков в balcony после обработки:',data['balcony'].isna().sum())
```

Обработаем пропуски в `ceiling_height`, заменяя их на медианное значение. Для корректной замены проверим есть ли зависимость высоты потолков от других характеристик квартиры

```python
#ceiling_height
data.corr()['ceiling_height']
```

Так как значение коэффициента корреляции Пирсона между высотой потолков и остальными характеристиками квартир ближе к нулевому значению,  пропуски в `ceiling_height` возможно заменить медианным значением столбца без учета влияния других характеристик.

```python
# заменим пропуски в ceiling_height на медиану 
mediana = data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(mediana)
#проверим , что пропусков не осталось 
print('Количество пропусков в ceiling_height после обработки:',data['ceiling_height'].isna().sum())
```

Обработаем пропуски в `kitchen_area`. 

Для этого посчитаем отношение общей площади к площади кухни по непропущенным значениям датасета в разрезе группировки по количеству комнат. 


```python
print('Количество пропусков до замены:',data['kitchen_area'].isna().sum())
print('Коэффициенты замены:')
# рассчитаем средний коэффициент отношения площади кухни к общей площади в разрезе количества комнат 
#и применим его к столбцу total_area, заменяя их произведением  пропуски в kitchen_area
for room in data['rooms'].sort_values().unique():
    ratio_kitch_total_area = (data.loc[data['rooms'] == room, 'kitchen_area']).median()/(data.loc[data['rooms'] == room, 'total_area']).median()
    print('Для',room ,'ком.-','{:,.2}'.format(ratio_kitch_total_area))
    data.loc[(data['kitchen_area'].isna())&(data['rooms'] == room), 'kitchen_area'] = ratio_kitch_total_area*data['total_area']
             
#проверим , что все пропуски заполнены
print('Количество пропусков после замены:',data['kitchen_area'].isna().sum())
```

Обработаем пропуски в `living_area`. 

 


```python
print('Количество пропусков до замены:',data['living_area'].isna().sum())
print('Коэффициенты замены:')
# рассчитаем средний коэффициент отношения площади кухни к общей площади в разрезе количества комнат 
#и применим его к столбцу total_area, заменяя их произведением  пропуски в kitchen_area
for room in data['rooms'].sort_values().unique():
    ratio_living_total_area = (data.loc[data['rooms'] == room, 'living_area']).median()/(data.loc[data['rooms'] == room, 'total_area']).median()
    print('Для',room ,'ком.-','{:,.2}'.format(ratio_living_total_area))
    data.loc[(data['living_area'].isna())&(data['rooms'] == room), 'living_area'] = ratio_living_total_area*data['total_area']
             
#проверим , что все пропуски заполнены
print('Количество пропусков после замены:',data['living_area'].isna().sum())
```

Обработаем пропуски в `days_exposition`.

Так как изначально в данном столбце нет нулевых значений, заполним пропуски в days_exposition нулями,обозначающими отсутствие информации о сроке размещения.

```python
#заменим Nan в days_exposition на нулевые значения
data['days_exposition'] = data['days_exposition'].fillna(0)
#проверим, что пропусков не осталось
print('Количество пропусков после замены:',data['days_exposition'].isna().sum())
```

Обработаем пропуски в `airports_nearest`,`cityCenters_nearest`,`parks_around3000`,`parks_nearest`,`ponds_around3000`,`ponds_nearest`,заменяя их служебным значением "999999", означающим отсутствие данных.



```python
# заменим Nan на служебное значение 999999
data = data.fillna({'airports_nearest':999999,
                     'cityCenters_nearest':999999,
                     'parks_around3000':999999,
                     'parks_nearest':999999,
                     'ponds_around3000':999999,
                     'ponds_nearest':999999})
#проверим количество пропусков после всех замен
data.isna().sum()
```

Проверим есль ли явные дубликаты в датасете



```python
data.duplicated().sum()
```

## Расчёты и добавление результатов в таблицу

Добавим в таблицу информацию 
 - о стоимости квадратного метра, 
 - сотношения жилой и общей площади,
 - отношении площади кухни к общей .



```python
# посчитаем  стоимость квадратного метра, поместив значения в новый столбец price_sq_meter
data['price_sq_meter'] = data['last_price']/data['total_area']
# посчитаем отношение жилой к общей площади, поместив значения в новый столбец living_to_total_area
data['living_to_total_area'] = data['living_area']/data['total_area']
# посчитаем отношение площади кухни к общей , поместив значения в новый столбец kitchen_to_total_area
data['kitchen_to_total_area'] = data['kitchen_area']/data['total_area']
# выведем обновленный датасет
data.head()
```

Добавим отдельные столбцы с днем недели, месяцем и годом публикации объявления.




```python
# применим функцию DatetimeIndex к weekday_exposition
data['weekday_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
# применим функцию DatetimeIndex к month_exposition
data['month_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).month
# применим функцию DatetimeIndex к year_exposition
data['year_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).year
# выведем обновленный датасет
data.head()
```

Проведем категоризацию квартир по этажам, выделив первый,последний и другой. 




```python
# напишем функцию, проставляющую дополнительный индикатор этажа  первый,  последний, другой
def floors_categ(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
# применим функцию floors_categ к датасету , записав результат в новый столбец floors_categ   
data['floors_categ']  = data.apply(floors_categ,axis=1)
# выведем обновленный датасет
data.head()
```

### Изучение 10 городов с наибольшим количеством объявлений


Исследуем 10 населенных пунктов с наибольшим чилом объявлений.

```python
#создадим таблицу со списком 10 городов с наибольшим количеством объявлений
top_10_locality_name = data['locality_name'].value_counts().sort_values(ascending=False).head(10).to_frame()
#найдем медианную стоимость за 1 кв.м. жилья для 10 городов, с наибольшим количеством объявлений
top_10_locality_name_sqrtmeter = (
                                    data.query('locality_name in @top_10_locality_name.index')
                                        .pivot_table(index='locality_name',values='price_sq_meter',aggfunc='mean')
                                        .sort_values(by='price_sq_meter',ascending=False)
                                        .reset_index(drop=False)
                                )      
# переименуем столбцы полученной таблицы
top_10_locality_name_sqrtmeter.columns = [['Город','Средняя цена за 1 кв.м.,руб.']]
#выведем 10 городов с их медианным значением за 1 кв.м.                                  
display(top_10_locality_name_sqrtmeter)
```

### Исследовательский анализ сегмента квартир в центре Санкт-Петербурга

```python
#создадим столбец с растоянием до центра в км - cityCenters_nearest_km
data['cityCenters_nearest_km'] = (data['cityCenters_nearest']/1000)
#округлим значения cityCenters_nearest_km до целого числа в меньшую сторону
data['cityCenters_nearest_km'] = data['cityCenters_nearest_km'].astype(int)

#построим график зависимости средней цены от расстояния до центра в км,
#используя срез данных без значений Nan в столбце cityCenters_nearest(были приравнены 999999) и записи только для Санкт-Петербурга
(
      data.query('cityCenters_nearest < 999999 and locality_name == "Санкт-Петербург"')
      .pivot_table(index='cityCenters_nearest_km',values='last_price',aggfunc='median')
      .plot(style='o-',grid=True,figsize=(10,8),kind='bar')
)
plt.show()
```

На графике видна явная разница в цене продажи между 8 и 9 км от центра Петербурга. 
Сегмент квартир в центре Санкт-Петербурга ограничим радиусом в 8 км включительно. 

Проанализируем данный сегмент квартир, изучив площадь, цену, число комнат и высоту потолков.

```python
#присвоем срезу данных для сегмента квартир в центре название
#data_centre_spb = data.query('locality_name == "Санкт-Петербург" and cityCenters_nearest_km < 9')
data_centre_spb = data[(data['locality_name'] == 'Санкт-Петербург')&(data['cityCenters_nearest_km'] < 9)]
# выведем статистики для сегмента квартир в центре Спб  
display(data_centre_spb[['last_price','total_area','rooms','ceiling_height','price_sq_meter']].describe())
#создадим список параметров, для которых будем строить графики
names = ['last_price','total_area','rooms','ceiling_height']
# с помощью цикла и функции hist нарисуем графики размаха 
for name in names:
    plt.figure(figsize=(15,10))
    ax = data[name].hist(bins=40,label='Весь рынок')
    data_centre_spb[name].hist(bins=40,ax=ax,alpha=0.6,label='Центр' )
    plt.legend()
    plt.title(name)
    plt.show()
```

В сегменте квартир в центре Санкт-Петербурга больше всего 2-комнатных и 3-комнатных квартир, в то время как на всем рынке Санкт-Петербурга самое востребованное предложение - это однокомнатные квартиры.

Доля квартир с площадью свыше 100 кв.м. от общего объема предложений выше в центре Санкт-Петербурга, чем на всем рынке Санкт-Петербурга.
Распределение квартир по цене также отличается : для центра характерны более дорогие квартиры , половина покупок приходиться на диапозон от 6 до 12 млн., средняя цена покупки - 9 млн.руб.В то время как для всего рынка половина проданных квартир приходится на диапозон от 3 до 6 млн .руб.

Доля квартир с высотой потолка от 3 метров выше в центральном сегменте , но для обоих рынков самая типичная высота квартир  - 2,65м.


```python
# рассчитаем коэффициент Пирсона для цены от площади квартиры, от числа комнат и от удаленности от центра
print('Зависимость цены от площади:',data_centre_spb['last_price'].corr(data_centre_spb['total_area']))
print('Зависимость цены от количества комнат:',data_centre_spb['last_price'].corr(data_centre_spb['rooms']))
print('Зависимость цены от удаленности от центра:',data_centre_spb.query('cityCenters_nearest<999999')['last_price'].corr(data_centre_spb['cityCenters_nearest']))

#построим диаграмму рассеивания для для цены от площади квартиры, от числа комнат и от удаленности от центра
names = ['total_area','rooms','cityCenters_nearest']
for name in names:
    if name == 'cityCenters_nearest':
        (
            data_centre_spb.query('cityCenters_nearest<999999')
            .plot(x=name, y='last_price', kind='hexbin', gridsize=50, figsize=(12,7), sharex=False, grid=True)
        )
        plt.show()
        (
            data.query('cityCenters_nearest<999999')
            .plot(x=name, y='last_price', kind='hexbin', gridsize=50, figsize=(12,7), sharex=False, grid=True)
        )
        plt.show()
    else:
        data_centre_spb.plot(x=name, y='last_price', kind='hexbin', gridsize=50, figsize=(12,7), sharex=False, grid=True)
        plt.title('Центр')
        plt.show()
        data.plot(x=name, y='last_price', kind='hexbin', gridsize=50, figsize=(12,7), sharex=False, grid=True)
        plt.title('Весь рынок')
        plt.show()

#посмотрим зависимость медианной рыночной стоимости от количества комнат

ax=(
    data_centre_spb
    .pivot_table(index='rooms',values='last_price',aggfunc='median')
    . plot(style='o-',grid=True)
)
(
    data
    .pivot_table(index='rooms',values='last_price',aggfunc='median')
    . plot(style='o-',grid=True,legend='Весь рынок',figsize=(10,5),ax=ax)
)
plt.legend(['Центр', 'Весь рынок'])
plt.title('Зависимость медианной стоимости квартиры от количества комнат')
plt.show()

```

Медианная стоимость для квартир с  одинаковым количеством комнат выше для сегмента квартир в центре, но квартиры свыше 5 комнат имеют практически одинаковую медианную стоимость.

```python
#посмотрим зависимость медианной рыночной стоимости от даты размещения
#создадим список столбцов для построения графиков циклом
names = ['weekday_exposition','month_exposition','year_exposition']
#напишем цикл для построения графиков зависимости медианной цены от дня/месяца/года размещения
for name in names:
    ax=(
        data_centre_spb
       .pivot_table(index=name,values='last_price',aggfunc='median')# графики для центрального сегмента квартир
       . plot(style='o-',grid=True,figsize=(10,6))
    )
    (
        data
       .pivot_table(index=name,values='last_price',aggfunc='median')#графики для всего рынка недвижимости Спб
       . plot(style='o-',grid=True,figsize=(10,6),ax=ax)
    )
    plt.legend(['Центр','Весь рынок'])
    plt.show()
#построим график средней удаленности от центра продаваемых квартир в разрезе года продажи
(
    data_centre_spb.query('cityCenters_nearest<999999')
    .pivot_table(index='year_exposition',values='cityCenters_nearest',aggfunc='median')
    .plot(style='o-',grid=True,figsize=(7,6))
)
plt.show()
```
